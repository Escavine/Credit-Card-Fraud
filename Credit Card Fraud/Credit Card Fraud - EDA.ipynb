### We can see that some of the data (i.e. V1, V2 etc.) haven't been processed properly and the only understandable pieces of data are 'Amount' and 'Class'

```python
# Finding a relationship between class and amount frauded

plt.figure(figsize=(10,6))
plt.scatter(df['Amount'], df['Class'])
plt.xlabel('Amount')
plt.ylabel('Class')
plt.title('Relationship between amount and class')
plt.show()
```

### From seeing the scatter graph with matplotlib, it is clear that within the transactions that have occured within 2 days, fraud has only occured a few times (492), and are amounts lower than 5000

### We can also tell that there is a huge imbalance with this dataset as the majority of cases are non-fraudulant (larger class) and only a few are fraudulant (minority class)

```python
df.hist(figsize=(20,20))
plt.show()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(16,12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Finding a relationship between amount and time

plt.figure(figsize=(10,6))

colors = {0: 'blue', 1: 'red'}

for cls, color in colors.items():
    subset = df[df['Class'] == cls]
    plt.bar(subset['Amount'], subset['Time'], 
            label=f'Class {cls}', color=color)
    
plt.xlabel('Amount')
plt.ylabel('Time')
plt.title('Relationship between Amount and Time')
plt.legend(title='Class')
plt.show()
```

### From the relationship displayed within the bar graph, we can see that the frauded amounts are lower than 5000 euros, and the longest frauded transaction is around 80,000 seconds (22 hours)
